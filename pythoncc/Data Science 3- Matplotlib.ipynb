```python
days=[1,2,3,4,5]
max_t=[50,55,40,33,44]
min_t=[2,3,4,14,2]
avg_t=[22,28,30,29,32]

plt.plot(days,min_t,label='min')
plt.plot(days,max_t,label='max')
plt.plot(days,avg_t,label='avg')

plt.xlabel("DAYS")
plt.ylabel("TEMPERATURE")
plt.title("DAYS VS TEMPERATURE")
plt.legend(loc="upper right")
plt.grid()
```

## Bar charts
- Bar char is basically useed for see various parameters

```python
company=["TCS","INFY","LTI","PSL"]
revenue=[400,300,200,160]
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
plt.bar(company,revenue,label="revenue")
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Graph of Company vs Revenue")
plt.legend()
```

```python
profit=[23,45,60,90]
```

```python
#plt.bar can gives vertical graph
```

```python
plt.bar(company,revenue,label="revenue")
plt.bar(company,profit,label="Profit")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.legend()
```

```python
# plt.barh can plot horizontal graph
```

```python
plt.barh(company,revenue,label="revenue")
plt.barh(company,profit,label="Profit",color="BLACK")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.legend()
```

# Pie charts 
- It use to plot the graph with % for each criteria for eg. company expences


```python
company_expences=[100,60,40,20,80]
expense_data=["Emp_salary","Infra","Food_exp","Traveling_Allow","Prizes"]
```

```python
plt.pie(company_expences,labels=expense_data,shadow=True,autopct='%0.1f%%',explode=[.1,0,0,0,0],radius=3)
```

# Histogram
- It is use to plot freqenecy

```python
blood_Sugar_men=[70,108,100,120,87]
blood_Sugar_women=[85,111,100,128,80]
blood_Sugar_child=[80,89,90,80,102]
```

```python
#plt.bar(blood_Sugar_men,blood_Sugar_women,label="blood_Sugar_men")
plt.hist([blood_Sugar_men,blood_Sugar_women,blood_Sugar_child],rwidth=9.5,label=['mens','womwns','CHILD'],color=['ORANGE','BLUE','GREEN'])
plt.xlabel("Quqntity")
plt.ylabel("Sugar level")
plt.title("SUGAR LEVEL")
plt.legend()
```
